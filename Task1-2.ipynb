#**1.Running Sum of 1d Array.**

```python
lst = list(map(int, input("Enter numbers: ").split()))
for i in range(1,len(lst)):
     lst[i]=lst[i]+lst[i-1]

print(lst)
```

#**2.Shuffle the Array.**

```python
lst=list(map(int,input("Enter numbers: " ).split()))
n=int(input("Enter a number: ").strip())
arr=[]
k=len(lst)//2
for i in range(k):
    arr.append(lst[i])
    arr.append(lst[k+i])

print(arr)
```

#**3.Kids With the Greatest Number of Candies.**

```python
candies=list(map(int,input("Enter number of candies: " ).split()))
extracandies=int(input("Enter number of extra candies: ").strip())
arr=[]
max_candies=max(candies)
for i in range(len(candies)):
     if(candies[i]+extracandies>=max_candies):
         arr.append(True)
     else:
         arr.append(False)
        
print(arr)
```
