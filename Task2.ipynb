**1.Defanging an IP Address**
Example 1:
Input: address = "1.1.1.1"
Output: "1[.]1[.]1[.]1"

```python
address= input().strip()
output="";
for i in range(0,len(address)):
    if(address[i]=="."):
        output=output+"[.]";
    else:
        output=output+address[i];
        
print (output)
```

**2.Find Numbers with Even Number of Digits**
Example 1:
Input: nums = [12,345,2,6,7896]
Output: 2

```python
nums = list(map(int, input().split(",")))
even=0
for val in nums:
    if len(str(val))%2==0:
        even=even+1
print(even)
```

**3.Number of Good Pairs**
Example 1:
Input: nums = [1,2,3,1,1,3]
Output: 4

```python
nums = list(map(int, input().split(",")))
count=0
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]==nums[j]:
            count=count+1
            
print(count)
```

**4.How Many Numbers Are Smaller Than the Current Number**
Example 1:
Input: nums = [8,1,2,2,3]
Output: [4,0,1,1,3]

```python
nums=list(map(int,input().split(",")))
c=[0]*len(nums)
for i in range(len(nums)):
    for j in range(len(nums)):
        if(nums[j]<nums[i]):
            c[i]=c[i]+1
            
print(c)
```

**5.Subtract the Product and Sum of Digits of an Integer**
Example 1:
Input: n = 234
Output: 15 

```python
n=int(input())
prod=1
s=0
while n > 0:
    c = n % 10
    prod *= c
    s += c
    n //= 10
        
print(prod - s)
```

**6.XOR Operation in an Array**
Example 1:
Input: n = 5, start = 0
Output: 8

```python
n=int(input())
start=int(input())
result=0
for i in range(n):
    result=result^(start+2*i)

print(result)
```
